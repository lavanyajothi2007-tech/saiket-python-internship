```python
# Currency Converter: INR to USD

amount_in_inr = float(input("Enter amount in INR: "))

# Fixed conversion rate (example)
conversion_rate = 0.012

amount_in_usd = amount_in_inr * conversion_rate

print("Converted Amount:")
print(f"INR {amount_in_inr} = USD {amount_in_usd:.2f}")
```
